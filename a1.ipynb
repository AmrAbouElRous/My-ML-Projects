```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x_train = np.array([1,2])
y_train = np.array([300,500])
print(f"x_train = {x_train}")
print(f"y_train={y_train}")
```

```python
print(x_train.shape)
m = x_train.shape[0]
print(f"Number of training examples is {m}")
```

```python
plt.scatter(x_train,y_train,marker="x",color="r")
plt.ylabel("price in 1000$")
plt.xlabel("size in 1000 square feet")
```

# Model
the model function for linear regression 
$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

<img align="left" src = "./Images/a13.png"  style=" width:500px; padding: 10px;">  <img>
</figure>

```python
def compute_model_output(x,w,b):
    """
    prediction of Linear model
    Args:
    x        (feature)
    w,b      (parameters)
    returns
    y        (target)
    """
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w* x[i] + b

    return f_wb
```

```python
plt.plot(x_train,compute_model_output(x_train,100,100),c="b",label="predicted value")
plt.scatter(x_train,y_train,marker="x",c="r",label ="actual value")
plt.ylabel("price in 1000$")
plt.xlabel("size in 1000 square feet")
plt.title("house pricing")
plt.legend()#plt.legend() is a function in Matplotlib that adds a legend to a plot, providing a visual key to identify different data series or elements on the plot
plt.show()
```
